```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
image = cv2.imread("Car.jpg")  #the imread function is used to read the images. 
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.title("Original")
plt.imshow(image_rgb)
```

```python
gray_image = cv2.imread("Car.jpg", cv2.COLOR_BGR2GRAY)
#Saving using imwrite()
cv2.imwrite("Car_gray.jpg", gray_image)

plt.title("Gray-Scale Mode")
plt.imshow(gray_image)
```

```python
#Loading 
img = cv2.imread("Car_gray.jpg")
#Displaying 
plt.title("Saved Image")
plt.imshow(img)
```
